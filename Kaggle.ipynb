```python
#we display the features
feature_data = pd.read_csv('kaggle_data/features.txt', header=None, sep="  ", names=['feature_names', 'feature_description'], engine='python')
feature_data.head(5)
```

```python
#we diplay the train set
list_feature_names = list(feature_data['feature_names'])
train_data = pd.read_csv('kaggle_data/train.csv', header=None, sep=" ", names=list_feature_names)
train_data.head(5)
```

```python
#display test set
test_data=pd.read_csv('kaggle_data/test-val.csv',header=None,sep=" ",names=list_feature_names)
test_data.head(5)
```

```python
#we display y the labels
target_data = pd.read_csv('kaggle_data/train-targets.csv', sep=",")
target_data.head(5)
```

```python
#Data engineering 
#drop irrelevant features (not now)

# 'weekday' encoding
weekday_data=pd.get_dummies(train_data['weekday'],prefix='weekday',drop_first=True)
other_data=train_data.drop(['weekday'],axis=1)
training_data=pd.concat([weekday_data,other_data],axis=1)

#print
training_data.head(5)
```

```python
#'category' encoding

category_data=pd.get_dummies(train_data['category'],prefix='category',drop_first=True)
other_data=train_data.drop(['category'],axis=1)
training_data=pd.concat([category_data,other_data],axis=1)

#print
training_data.head(5)

#From now we will us : training_data as our train set
```

```python
#Data standardization
from sklearn import preprocessing
sc=preprocessing.StandardScaler()
#X_scaled = preprocessing.scale(X)
```

```python
#we are going to load data X and y for regression
sc.fit(training_data)
X=sc.transform(training_data)
y=target_data.values

#data is ready for regression
```

```python
from sklearn import cross_validation
folds= cross_validation.KFold(y.size, n_folds=5, shuffle=True)
```

```python
def cross_validate_regr(design_matrix, labels, regressor, cv_folds):
    """ Perform a cross-validation and returns the predictions.
    
    Parameters:
    -----------
    design_matrix: (n_samples, n_features) np.array
        Design matrix for the experiment.
    labels: (n_samples, ) np.array
        Vector of labels.
    regressor:  Regressor instance; must have the following methods:
        - fit(X, y) to train the regressor on the data X, y
        - predict(X) to apply the trained regressor to the data X and return estimates 
    cv_folds: sklearn cross-validation object
        Cross-validation iterator.
        
    Returns:
    -------
    pred: (n_samples, ) np.array
        Vectors of predictions (same order as labels).
    """
    pred = np.zeros(labels.shape)
    for tr, te in cv_folds:
        regressor.fit(design_matrix[tr,:], labels[tr])
        pred[te] = (regressor.predict(design_matrix[te,:]))
    return pred
```

```python
from sklearn import linear_model
from sklearn import metrics

#operation linear regression
regr=linear_model.LinearRegression()
y_pred=cross_validate_regr(X,y,regressor=regr,cv_folds=folds)

#RMLSE (Root mean squared logarithmic error)
print("Root Mean squared log error: %.3f" % metrics.mean_squared_log_error(y, y_pred))
```
